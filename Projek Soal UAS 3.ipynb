```python
# Import libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

# Data
data = pd.read_excel("https://github.com/ekayuniar/datamining/raw/master/UAS/dataset%20soal%20no.3.xlsx")

# Print the first few rows of the data
print(data.head())

# Remove leading/trailing spaces in column names
data.columns = data.columns.str.strip()

print("Columns in data:", data.columns)  # Verifying column names

# Separate features and target
X = data[['Usia', 'Kelahiran ke-', 'Waktu Kelahiran', 'Tekanan Darah', 'Kelainan Jantung']]
y = data['Cessar']

# Verify lengths
print("Length of X:", len(X))
print("Length of y:", len(y))

# Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Define KNN model
knn = KNeighborsClassifier(n_neighbors=5)

# Train the model
knn.fit(X_scaled, y)

# Test data points
# Scenario 1: Ibu hamil dengan Usia 30 Tahun, Kelahiran ke-1, Waktu kelahiran sesuai HPL, Tekanan darah Normal
test_point_1 = np.array([[30, 1, 1, 0, 1]])
# Scenario 2: Ibu hamil dengan Usia 29 Tahun, Kelahiran ke-2, Waktu kelahiran sesuai HPL, Tekanan darah Tinggi
test_point_2 = np.array([[29, 2, 1, 1, 1]])

# Normalize test points
test_point_1_scaled = scaler.transform(test_point_1)
test_point_2_scaled = scaler.transform(test_point_2)

# Make predictions
prediction_1 = knn.predict(test_point_1_scaled)
prediction_2 = knn.predict(test_point_2_scaled)

print(f'Prediction for test point 1: {prediction_1[0]}')
print(f'Prediction for test point 2: {prediction_2[0]}')
```
